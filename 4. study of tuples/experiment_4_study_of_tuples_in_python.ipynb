```python
##Python tuple
#create and access tuple items
thistuple =("Mango","Orange","Blueberry")
print(thistuple[1])
```

```python
numbers = (3,7,1,9,4)
print(numbers(1::2))
```

```python
numbers = (10,20,30,40,50)
#Slicing examples
print("elements from index 1 to 3:")
print(numbers[1:3])
print("first 3 elements:")
print(numbers[0:3])
print("Elements from index 2 onwards:(reverse indexing)")
print(numbers[-3:])
print("all elements:")
print(numbers[0:6])
```

```python
marks = (78, 85, 90, 66, 88)
print("maximum marks: ",(marks[2]))
print("minimum marks: ",(marks[0]))
print("total marks: ",(sum(marks)))
print("average marks: ",(sum(marks)/len(marks)))
print("sort the marks in ascending order: ",(sorted(marks)))
```

```python
'''2. Grocery Shopping list
A user maintains a grocery list for monthly shopping.
tasks:
create a grocery list
add new items to the list
remove purchased items
display the final shopping list
concepts used - append(), remove(), indexing'''
Grocerylist = ("fruits", "vegetables", "milk", "oats")
print("Grocery list:", Grocerylist)
Grocerylist.append("juice")
print("the purchased items ", Grocerylist)
Grocerylist.remove("vegetables")
print("the final grocery list: ",Grocerylist)
```
