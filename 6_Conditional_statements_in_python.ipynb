```python
#7. Python Program to check if a date is valid and print the incremented date if it is valid
# Input date in dd-mm-yyyy format
date = input("Enter date (dd-mm-yyyy): ")

# Split the date
parts = date.split("-")

# Check correct format
if len(parts) != 3:
    print("Invalid date")
    exit()

day = int(parts[0])
month = int(parts[1])
year = int(parts[2])

# Check leap year
leap = False
if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    leap = True

# Find max days in the month
if month == 1 or month == 3 or month == 5 or month == 7 or month == 8 or month == 10 or month == 12:
    max_days = 31
elif month == 4 or month == 6 or month == 9 or month == 11:
    max_days = 30
elif month == 2:
    if leap:
        max_days = 29
    else:
        max_days = 28
else:
    print("Invalid date")
    exit()

# Validate day
if day < 1 or day > max_days:
    print("Invalid date")
    exit()

# Increment date
day += 1

if day > max_days:
    day = 1
    month += 1

    if month > 12:
        month = 1
        year += 1

# Print incremented date
print("Incremented date:", f"{day:02d}-{month:02d}-{year}")
```

```python
#9. salary calculation with allowances
# write a python program to calculate the gross salary of an employee based on basic salary
# if basic <= 10000: HRA = 20%, DA = 80%
# if basic <= 20000: HRA = 25%, DA = 90%
#otherwise: HRA = 30%, DA = 95%
basic = float(input("enter basic salary: "))
if basic <= 10000:
  HRA = basic * 0.20
  DA = basic * 0.80
  gross = basic + HRA + DA
  print("the gross salary is", gross)
elif basic <= 20000:
  HRA = basic * 0.25
  DA = basic * 0.90
  gross = basic + HRA + DA
  print("the gross salary is", gross)
else:
  HRA = basic * 0.30
  DA = basic * 0.95
  gross = basic + HRA + DA
  print("the gross salary is", gross)
```

```python
#10. write python program to calculate income tax based on annual income:
income = float(input("Enter your annual income: "))
tax = 0
if income <= 250000:
  tax = 0
elif income <= 500000:
  tax = (income - 250000) * 0.05
elif income <= 1000000:
  tax = (250000 * 0.05) + (income - 500000) * 0.20
else:
  tax = (250000 * 0.05) + (500000 * 0.20) + (income - 1000000) * 0.30
print("income tax: ", tax)
```
