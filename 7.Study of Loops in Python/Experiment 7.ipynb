## the break statement
## with the break statement we can stop the loop even if the while condition is true:

```python
#exit the loop when i = 3:
i = 1
while i < 6:
   print(i)
   if i == 3:
     break
   i += 1
```

```python
#search for an element in a list
nums = [20,30,40,50,60]
key = int(input("enter element to search: "))
i = 0
while i<len(nums):
    if nums[i] == key:
      print("element found at index", i + 1)
      break
    i += 1
else:
  print("element not found")
```

##While- continue
### continue skips the remaining statements of the current iteration and jumps back to the while condition

```python
#print numbers from 1 to 10 (skip 5)
i = 1
while i <= 10:
  if i == 5:
      i += 1
      continue
  print(i)
  i += 1


```

```python
#print only odd numbers (1 to 10)
i = 1
while i <= 10:
  if i % 2 == 0:
      i += 1
      continue
  print(i)
  i += 1
```

```python
#print only even numbers (1 to 10)
i = 1
while i <= 10:
  if i % 2 != 0:
      i += 1
      continue
  print(i)
  i += 1
```
