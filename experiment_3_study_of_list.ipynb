```python
#Reverse/negative Indexing
mylist = ["strawberry","mango","Guava"]
print(mylist[-1])
```

```python
#Slicing
mylist = ["strawberry","mango","Guava","apple","cherry","banana"]
print(mylist[2:5])
print(mylist[:5])
print(mylist[3:])
#:4 first 4, 3: last 3
```

```python
mylist = ["strawberry","mango","Guava","apple","cherry","banana"]
print(mylist[:-2])
print(mylist[-3:-1])
```

```python
#change Item Value
mylist = ["apple", "banana", "cherry"]
mylist[1] = "blackcurrant"
print(mylist)
```

```python
#append items
mylist = ["apple", "banana", "cherry"]
mylist.append("orange")
print(mylist)
```

```python
#insert items
mylist = ["apple", "banana", "cherry"]
mylist.insert(1, "orange")
print(mylist)
```

```python
#extend list
mylist = ["apple", "banana", "cherry"]
tropical = ["mango", "pineapple", "papaya"]
mylist.extend(tropical)
print(mylist)
```

```python
numbers = [10,20,30,40,50]
#Slicing examples
print("elements from index 1 to 3:")
print(numbers[1:3])
print("first 3 elements:")
print(numbers[0:3])
print("Elements from index 2 onwards:(reverse indexing)")
print(numbers[-3:])
print("all elements:")
print(numbers[0:6])
```

```python
marks = [78, 85, 90, 66, 88]
print("maximum marks: ",(marks[2]))
print("minimum marks: ",(marks[0]))
print("total marks: ",(sum(marks)))
print("average marks: ",(sum(marks)/len(marks)))
print("sort the marks in ascending order: ",(sorted(marks)))
```

```python
'''2. Grocery Shopping list
A user maintains a grocery list for monthly shopping.
tasks:
create a grocery list
add new items to the list
remove purchased items
display the final shopping list
concepts used - append(), remove(), indexing'''
Grocerylist = [ "fruits", "vegetables", "milk", "oats"]
print("Grocery list:", Grocerylist)
Grocerylist.append("juice")
print("the purchased items ", Grocerylist)
Grocerylist.remove("vegetables")
print("the final grocery list: ",Grocerylist)
```

```python
contacts = ["Dad","Sachin","Riya","Fatima","Mom"]
print("phone contacts:",(contacts))
contacts.append("Abhi")
print("new contact list: ",(contacts))
contacts.remove("Riya")
print("updated contact list: ",(sorted(contacts)))
```

```python
#temperature analysis : daily temperature reading of a city are stored in a list for 10 days
Templist =[20,45,26,39,31,37,32,48,19,18]
print("daily temperature of first 5 days: ",Templist[0:5])
print("daily temperature of last 5 days:  ", Templist[5:10])
print("highest temperature: ",max(Templist))
print("lowest temperature: ", min(Templist))
print("average temperature: ", sum(Templist)/len(Templist))
```

```python
#Daily expense tracker
expense = [300,400,500,600,700,800,100,900]
print("store daily expenses: ", expense)
print("display total and average expenses: ", sum(expense), sum(expense)/len(expense))
print("highest spending day: ", max(expense))
print("lowest spending day: ", min(expense))
```
